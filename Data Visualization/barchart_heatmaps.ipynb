```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

flights_data = pd.read_csv('dataIn/flight_delays.csv', index_col='Month')

flights_data.head()
```

# Bar Chart

We can create a bar chart using the `sns.barplot()` function where we specifcy the `x` and `y` data for it.

```python
plt.figure(figsize=(14,6))
plt.title("Average Arrival Delay for Spirit Airlines Flights, by Month")
sns.barplot(x=flights_data.index, y=flights_data['NK'])
plt.ylabel("Arrival delay (in minutes)")
```

# Heatmaps

Very useful when looking for correlations or patterns. We can create one using `sns.heatmap()` an important parameter to know is `annot=True` which shows the numerical value in each cell, however for many variables (columns) things can get crowded.

In just is simple form a heatmap shows: for Month X how much delay do we have but it can be especially useful when used with correlation matrices.

```python
plt.figure(figsize=(14,6))
sns.heatmap(data=flights_data, annot=True)
```
