```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

iris_data = pd.read_csv('dataIn/iris.csv')
```

# Histograms 

We usually use histograms to view the distribution of our graph and see if it has a normal distribution or not. Especially when doing econometrics or BLUE OLS regressions we want to have a normal distribution and this graph can help use greatly. 

We can create one using `sns.histplot()`

```python
plt.figure(figsize=(14,6))
sns.histplot(iris_data['Petal Length (cm)'])
```

> As we can see the distribution is clearly not well distributed when it comes to Petal Lenght.

# KDE Plots

While they can be considered a smoothened histogram they are slightly different because a KDE plot loses the count of how many values fall into that certain category. A KDE plot's purpose is to show the distribution more neatly and easier to understand and is prefered when the specific count of each value is not a necessity.

We can create one using `sns.kdeplot()`. The `shade` property of the plot tells whether or not the area under the curved should be colored (usually you want it to be)

```python
plt.figure(figsize=(14,6))
# sns.kdeplot(data=iris_data['Petal Length (cm)'])
sns.kdeplot(data=iris_data['Petal Length (cm)'], shade=True)
```

> Not a normal distribution

# 2D KDE plots

We can also create a 2D KDE plot using a `joinplot()` where eachh axis represents a variable and its values and then the top of the graph is the distribution for axis x and the right side of the graph is the distribution for axis y.

> Note that `plt.figure()` wont work here because joinplot creates its own figure

```python
sns.jointplot(x=iris_data['Petal Length (cm)'], y=iris_data['Sepal Width (cm)'], kind="kde", shade=True)
```

## How to read a 2D KDE plot

A 2D KDE plot shows where combinations of two variables are more or less concentrated.

Inner contours usually indicate higher estimated density, meaning many observations are packed into that region.
Outer contours indicate lower density, where combinations are less common.
Each contour line connects points with the same estimated density, similar to elevation lines on a map.

So if a randomly chosen (x, y) pair falls inside a high-density region, that combination is more typical in the dataset than a pair in an outer low-density region.

In our dataset we can see that we have two separate clusters indicating two commonly dense regions in our data.

# Color-coded Histograms

We can create color-coded histograms that separate a variables distribution in relationship to another variable. For example we can see the distribution of petal length based on its species.

```python
sns.histplot(data=iris_data, x='Petal Length (cm)', hue='Species')
plt.title("Histogram of Petal Lengths, by Species")
```

We can do the exact sanme thing using a KDE plot.

```python
sns.kdeplot(data=iris_data, x='Petal Length (cm)', hue='Species', shade=True)
plt.title("Distribution of Petal Lengths, by Species")
```
