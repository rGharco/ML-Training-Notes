```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

spotify_data = pd.read_csv('dataIn/spotify.csv',index_col="Date", parse_dates=True) 
# -- We use parse_dates() to transform string dates into actual datetime objects
spotify_data.head()
```

# Seaborn | Matplotlib

Seaborn is a Python package that allows for fast and quick graphs to be created that are not verbose and are easy to understand. Because seaborn is built on matplotlib both of them work together the difference being that:
- seaborn does the styling part of the graph (styles, colors, types of graphs)
- matplotlib does the canvas rendering (dimensions, DPI, layout)

Internally when we say `plt.figure(figsize(10,10)))` matplotlib saves a new figure and sets it as the **current figure** which then seaborn designs as it pleases.

# Line charts

We can do line charts in seaborn using `sns.lineplot()`:

```python
plt.figure(figsize=(14,6))
plt.title("Daily Global Streams of Popular Songs in 2017-2018")
sns.lineplot(data=spotify_data)
```

## Plot a subset of the data

We can do so by selecting the column we want to create the graph for from the DataFrame.

```python
plt.figure(figsize=(14,6))
plt.title("Daily Global Streams of Despacito in 2017-2018")
sns.lineplot(data=spotify_data['Despacito'], label="Despacito")
plt.xlabel("Date")
```

## Plotting multiple line graphs

For multiple line graphs that we want interconnected or to see the difference in trends we can do create two lineplots like so:

```python
plt.figure(figsize=(14,6))
plt.title("Daily Global Streams of Popular Sngs in 2017-2018")
sns.lineplot(data=spotify_data['Despacito'], label="Despacito")
sns.lineplot(data=spotify_data['Shape of You'], label="Shape of You")
plt.xlabel("Date")
```
