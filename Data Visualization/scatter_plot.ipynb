```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

insurance_data = pd.read_csv('dataIn/insurance.csv')
```

# Scatter plots

They are used to showcase the **relationship between two variables** and you can create one using `sns.scatterplot()` specifying each axis with one variable. 

```python
plt.figure(figsize=(14,6))
plt.title("Link between BMI and insurance charges")
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

We can see in the graph above that a higher BMI leads to higher charges of insurance since higher BMI comes with risks of chronic health and heart disseases.

# Regression plot

We can use a regression plot to view the regression line which will showcase even clearer the relationship between the data. We can create one using `sns.regplot()`

```python
plt.figure(figsize=(14,6))
plt.title("Link between BMI and insurance charges")
sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

# Three variable scatter plots

You can add a third variable to the scatter plots to create visual **color-coded graphs** that are much more impactful. For example you can check if being a smoker affects the relationship between BMI and insurance costs. We can do so adding the `hue` attribute to the scatterplot.

```python
plt.figure(figsize=(14,6))
plt.title("Link between BMI and insurance charges")
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])
```

We can interpret this as follows:

> Having a high BMI AND being a smoker results in higher insurance charges while just having high BMI but NOT being a smoker leads to slightly lower insurance costs.

We can create two regression lines for both cases using the `sns.lmplot()`

```python
sns.lmplot(x="bmi", y="charges", hue="smoker", data=insurance_data)
```

> Note that instead of specifying the Series by doing insurance_data["bmi"] we put the data directly as paramter (data=insurance_data) and x, y, and hue just become the column names

# Swarmplot

When we have a catergorical variable and we want to see its link with another variable we use a swarmplot. That is because a sworplot instead of overlapping same values it spreads them apart and makes each of them visible (so if I have 9 yes answers with the same insurance costs then we get 9 spread out dots instead of 1 overlapping dot on the graph).
- this can help view outliers and frequency 
- used when we deal with categorical variables

We can create one using: `sns.swarmplot()` where we only specify x and y axis

```python
sns.swarmplot(x=insurance_data['smoker'], y=insurance_data['charges'])
```
