# DataFrame (pandas)

A DataFrame is a table-like structure specific to the pandas library. In order to create a DataFrame we can use the command `pd.DataFrame()` and supply the parameters necessary to create it. 

We can obtain information about a DataFrame using the `shape` attribute which returns us a `tuple` where the first value is the number of **rows** and the second attribute is the number of **columns**. Another way we can view information about a DataFrame is using the `head()` method which returns the first 5 rows of the DataFrame. We can also use the `info()` method which tells us how much memory the DataFrame occupies (in RAM), the Non-Null value count, the data types for each column and a sum of how many data types we have. Lastly we can use `describe()` to obtain basic statistical information about each column (mean, min, max, std deviation, etc.)

```python
df = pd.DataFrame(
    data={
        'Summer': [39.9, 37.9], 
        'Winter': [4.0,2.0]
    },
    index=["Max temperature", "Min Temperature"]
)

print(df)

print(df.shape)
print(type(df.shape))
print(df.head())
print(df.info())
print(df.describe())
```

As seen in the code above the `data` parameter is a dictionary where the key is the column name and the value is a `list` with each value mapped to a row (**note that any iterable can be used as data here**). The `index` parameter represents the label given to each row. 

## pandas Series

A Series is kind of like a column in a DataFrame, in fact you can say a DataFrame is composed of more Series "glued together". You can construct a Series by only using a list.

```python
print("----- Series ------")
print(pd.Series(
    data=[39.0, 37.0, 30.0],
    index=["Max Temperature", "Mean Temperature", "Min Temperature"],
    name="Temperatures Summer"
))

series = pd.Series(
    data=[39.0, 37.0, 30.0],
    index=["Max Temperature", "Mean Temperature", "Min Temperature"],
    name="Temperatures Summer"
)

print("\n----- DataFrame ------")
df = series.to_frame()
print(df)
```

## Reading DataFrames

Pandas allows for a variety of file extensions to be read but most notably **CSV (Comma Separated Values)** file types are the most used ones. We can read such files using `pd.read_csv()`. 

```python
df = pd.read_csv('dataIn/world_gdp_data.csv', encoding='latin1')

print(df.head())
print(df.info())
print(df.describe())
print(df.shape)
```

As we can see in the above dataset we have an `encoding` parameter set to our DataFrame, that is becaause the original dataset did not use UTF-8 encoding therefore we can speciy the encoding using that parameter. On top of this we created some descriptions for our data but we can barely see it, we can also make the column with the country name be the row index by specifying the `index_col` attribute.

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)

df = pd.read_csv('dataIn/world_gdp_data.csv', encoding='latin1', index_col=0)

print(df.head())
print(df.info())
print(df.describe())
print(df.shape)
```

The two options set with `pd.set_option()` allow us to view the whole dataframe and columns instead of a truncated output. However they are more useful when working with an entire terminal so they can be seen clearer.
