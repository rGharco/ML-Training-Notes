# Indexing, Selecting & Assinging

## Native Accesors

Since a DataFrame is an Object itself we can access a column using the property of such object so for example in our dataset we can do `df.indicator_name` and we can access its values (row value).

A DataFrame works also as a dictionary and we can access its values just like we would with any dictionary `df["indicator_name"]`. The difference between the two is that if a column name contains spaces `indicator name` we could not use native accessors but instead we would be forced to use the dictionary accessing method.

## Indexing

Selecting values in pandas can be done using `iloc` and `loc`. The difference between the two is what parameters you supply to each.

### iloc

iloc works with **indexes** with row first and column second so if you want to select row 1 column 1 you would do `df.iloc[0,0]`. This is indexing works with pythons `stdlib` so you can also use slicing operations on it like `df.iloc[:3,:]` to select all rows until the thrid and all columns. 

You can also use a `list` to index with `iloc` so for example `df.iloc[[1,2,3,5]]` will work.

### loc

loc works with **labels** and is particularely helpful when the indexes (row names) are strings. For that we will modify the dataframe `read_csv(index_col = 0)` and we will use the country names with `loc`. On top of that loc works with column names instead of their indexes so `df.loc["Afhanistan", "indicator_name"]` will work just fine.

> iloc uses python stdlib indexing so 0:10 will have values from 0..9 while loc uses indexing that takes 0:10 to be 0..10 included.

```python
print(df.iloc[0,0])
print(df.iloc[:3,:])

df = pd.read_csv('dataIn/world_gdp_data.csv', encoding='latin1', index_col=0)

print("--- loc ---")
print(df.loc["Afghanistan",:])
```

## Selecting

To filter out dataset and SELECT (like in SQL) we can use the comparison operators already known (==,<,>). This operation on a column in a DataFrame will return a Series that has values True/False where that condition is satisfied. We can then use this Series for further filtering like for example `df = df[df["1980] > 5]` which selects only the countries that had in 1980 a GDP growth over 5%.

We can also filter values in a more pretty way by using `loc` and saying `df.loc[df["1980"] > 5]` which does the exact same thing as the filtering statement above.

### Piping 

You can use `&` or `|` to combine operations (and something, or something). So for example `df.loc[(df["1980"] > 5) | (df["1989"] < 3])`.

### isin, isnull, isnotnull

`isin()` helps you select values that are inside that interval so for example `df.loc[df["1980"].isin([2.5,2.3])]` so all countries with GDP growth in 1980 that is either 2.5 or 2.3. 

`isnull()` for example you can use this to see NaN values inside your data (missing values) `df.loc[df["1980"].isnull()]`. And `isnotnull()` does the opposite.

```python
print(df['1982'] > 5)

print("--- SELECTING ---")
print(f"Shape before filtering: {df.shape}\n")

df = df[df["1980"] > 5]
print(f"Shape after filtering: {df.shape}\n")

df = df.loc[df["1980"] > 5]
print(f"Shape after loc operation: {df.shape}")

df = df.loc[(df["1980"] > 5) | (df["1989"] < 3)]
print(f"Shape after piping OR: {df.shape}")

df = df.loc[(df["1980"] > 5) & (df["1989"] < 3)]
print(f"Shape after piping AND: {df.shape}")

print("Null values")
print(df.loc[df["1980"].isnull()])
```

## Assigning

To assign data it is pretty simple just use the `=` operator on a column or row so `df["1980"] = 2.5` will make all values for that column to be 2.5. You can combine this with `loc` or `iloc` to assign values dinamically so `df.iloc["Afghanistan", "1980"] = 2.5` will assign 2.5 to the country (row) Afghanistan on column with year 1980.

```python
df = pd.read_csv('dataIn/world_gdp_data.csv', encoding='latin1', index_col=0)

print(df.loc["Afghanistan", "1980"])

df.loc["Afghanistan", "1980"] = 2.5

print(df.loc["Afghanistan", "1980"])
```
