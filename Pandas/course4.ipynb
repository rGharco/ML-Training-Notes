```python
import pandas as pd

weather_dataset = pd.read_csv("dataIn/weatherHistory.csv")
weather_dataset.head()
```

# Grouping 

When we say grouping we mean to literally combine values that are repetitive into groups (so for example the group with the same Country). In pandas much like in SQL we can do so using the `groupby()` method. As we can see in the example below we can see the maximum temperature for each weather type (breezy, clear, drizzle etc.) using a groupby.

> Note that `groupby()` returns a groupby object which has to have operations made on it, as of that moment after grouping you will have access to all the remaining columns to work with. So in theory it works just like a DataFrame meaning you can apply methods such as `apply()` to it.

You can also groupby multiple columns such as the `Summary` and the `Precip Type` like in the example below. This creates a **Multi-index** which is harder to manage and select data with.

> **DO NOT FORGET** that in order to work with columns resulted by using groupby you must use either an accessor `[]` or `iloc, loc` otherwise operations will be done on the entire DataFrame (max, mean etc.).

```python
print(type(weather_dataset.groupby(by='Summary')["Temperature (C)"].max()))
# print(weather_dataset.groupby(by='Summary')["Temperature (C)"].max())

print("---")

# print(weather_dataset.groupby(by='Summary').apply(lambda df : df["Temperature (C)"].max()))

print(weather_dataset.groupby(by=["Summary", "Precip Type"])["Temperature (C)"].sum())
```

## agg()

You can use the `agg()` function on a groupby object to apply multiple functions to the same groupby object instead of just one, kind of like a `describe()`.

```python
weather_dataset.groupby(by="Summary")["Temperature (C)"].agg(["min", "max", "mean"])
```

## Multi-indexes 

Because multi-indexes are hard to deal with and usually require much more complex manipulation your usual move would be to revert back to a single index using the `reset_index()` function.

```python
weather_dataset = pd.read_csv("dataIn/weatherHistory.csv")

grouped = weather_dataset.groupby(by=["Summary", "Precip Type"])["Temperature (C)"].sum()
type(grouped.index) # -- Multi-index
 
grouped = grouped.reset_index() # -- Ranged index
type(grouped.index)
```

# Sorting

Here you have two options:
- sort by value -> `sort_values()`
- sort by index -> `sort_index()` 

Both which are pretty self explanatory

```python
weather_dataset["Temperature (C)"].sort_values(ascending=False).head() # - sort descendingly
# weather_dataset.sort_values(by="Temperature (C)", ascending=False) # - same thing as above bt returns the whole dataframe sorted rather than only a series

weather_dataset.sort_index(ascending=False)
```
