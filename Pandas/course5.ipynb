```python
import pandas as pd

weather_dataset = pd.read_csv('dataIn/weatherHistory.csv')
weather_dataset.head()
```

# Dtypes

Data types in pandas are represented as `dtypes` and we can view them for each column using `weather_dataset.dtypes`. 

- int -> dtype int64
- float -> dtype float64
- string -> object 
- missing values -> NaN (Not a Number) - for technical reasons it is a float64 


# astype()

We can switch from one dtype to the other using `astype()`

```python
weather_dataset["Temperature (C)"].astype('int64') 
```

# Filling NaN values

We can use the `fillna()` function to fill them. On top of this we have the `pd.isnull()`and `pd.notnull()` for returning the NaN values in the DataFrame.

```python
pd.isnull(weather_dataset)
weather_dataset["Apparent Temperature (C)"].fillna(weather_dataset["Apparent Temperature (C)"].mean())
```

# Replacing values

You can replace values inside a DataFrame using the `replace()` method. This can be useful if you want for example to turn values from `0` to `Unknown` or something else. 

```python
weather_dataset["Temperature (C)"].replace(0, "Unknown")
```
