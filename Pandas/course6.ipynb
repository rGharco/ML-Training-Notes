```python
import pandas as pd

weather_dataset = pd.read_csv('dataIn/weatherHistory.csv')
```

# Renaming

In pandas you can rename both indexes and columns using the `rename()` function that can take as parameters either index or columns. Although you wont usually rename indexes, it is possible.

```python
# weather_dataset.rename(index={0: "Zero", 1: "One"}) # - uncomment to see change in indexes
weather_dataset.rename(columns={"Summary" : "Weather in that day"})
```

## Renaming axis

As you remember from `course.ipynb` both columns and row have a `name` attribute that can be set to them and we can rename that attribute using `rename_axis()` function

```python
weather_dataset.rename_axis("entry nr", axis="index").rename_axis("features", axis="columns")
```

# Combining DataFrames

- simple mushing together -> `pd.concat()`
- combining based on criteria -> `df.join()`

`merge()` is also an option but `join()` can usually do the same thing as it does.

```python
world_gdp_dataset = pd.read_csv('dataIn/world_gdp_data.csv', encoding='latin1')

pd.concat([weather_dataset, world_gdp_dataset])
```
